J'ai vérifié précédemment que les valeurs vides, dans les trois tables, sont des Nan et pas des espaces divers.

### Table 1 : caract-2024

```python
df = pd.read_csv('https://static.data.gouv.fr/resources/bases-de-donnees-annuelles-des-accidents-corporels-de-la-circulation-routiere-annees-de-2005-a-2024/20251021-115900/caract-2024.csv', encoding="utf-8", sep=';', engine='python')
```

```python
nb = df.isna().any(axis=1).sum()
print(nb)
```

```python
nb = df.drop(columns=['adr']).isna().any(axis=1).sum()
print(nb)
```

Ce sont les colonnes 'adr' qui sont vides -> ça n'est pas grave pour ce qu'on veut faire (on ne va pas utiliser l'adresse précise dans notre analyse). On peut garder toutes les lignes du dataset.
